Some terminology related to Data:
- **Transformation** - we apply mathematical operataions to change the distribution or scale of data
    - Data distribution or data scale
    
- **Normalization** - we adjust the values to a common scale without distorting the range - so that no features of data dominate the others due to scale
    - when we want our data to have bounds - upper and lower
    - Range = 0 to 1

- **Standardization** - convert data into a standard form 
    - rescale the features such that the mean comes out to be zero
