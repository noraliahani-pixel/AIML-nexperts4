#### **String Operations**

```python
sentence = "    What a wonderful pretty world!    "
```

```python
print(sentence.upper()) # 
print(sentence.lower()) # when user passes an input, always lower it first
print(sentence.title()) # capitalizes strating alphabets

print(sentence.strip()) # removed leading and trailing empty spaces
print(sentence.rstrip()) # remove trailing spaces
print(sentence.lstrip()) # remove leading spaces
```

```python
print(sentence.find("w")) # find the index of first matching character
print(sentence.find("world")) # find the index of first matching word

print(sentence.index("w")) # find the index of the first character of

print(sentence.find("ugly")) # -1 ----> is not found

print(sentence.count("w"))

print(sentence.replace("What","How is this"))
print(sentence.replace(" ","")) # replace empty space with, no space 
#---> clean the text of whitespaces
```

### **List**
- A collection of items
- Think of it as: a row of labelled boxes
- we create a list using square brackets `[]`

```python
# use a LIST
roll_numbers = [10,22,9,8]

type(roll_numbers)
print(roll_numbers)
```

```python
# access an element within a variable or a list
roll_numbers[0] # get me the first element
roll_numbers[1] # get me the second element
student4 = roll_numbers[3] # get the element at fourth position
```

If you try to access something that is not within the range of this list:
- IndexError: list index out of range 

```python
# go to third roll number, and change it 
roll_numbers[2] = 99
# we assigned a new value

roll_numbers
```

### **Tuple**
- A collection of items
- Are Immutable - Not changeable
- we create a tuple using round brackets `()`

```python
# use a tuple, save things you do not want to change
registration_nums = (101, 397, 172, 258)
```

```python
registration_nums[0]
registration_nums[3]

# can we change the registration number of the 4th student?
registration_nums[3] = 754

# TypeError ---> tuple does not allow reassignment 
```
