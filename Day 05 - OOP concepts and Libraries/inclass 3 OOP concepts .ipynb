### 1. Encapsulation - 
- Keeping together the data and the methods that work on that data
- Hiding the inner details
- 
- Ḍata should be changed through methods only

````python
class Animal:
    
    def __init__(self, name,age, species):
        self.name = name
        self.species = species
        
        # we want to keep as hidden
        # later on other programmers see this code, they will deduce that
        # Don't change this attribute directly from outside
        self._age = age

    # encapsulation idea is ---> not touching _age directly
    # but using getter and setter methods
    
    # getter methods --- Age
    def get_age(self):
        return self._age
    
    # getter method --- Species
    def get_species(self):
        return self.species

    # setter methods
    def set_age(self, new_age):
        # one could keep checks here as well like, 
        # example: not allowing negative nubmers for age
        if new_age > 0:
            self._age = new_age
        else:
            print("Age cannot be negative")
        
        return f"The new age is {new_age}"
    
    def introduce(self):
        print(f"Hello! I am ---> {self.name}.\n---> I am a {self.species} and I am {self._age} years old.")
````

````python
pet1 = Animal("bruno", 4, "koala")
````

````python
pet1.get_species()
````

````python
pet1.get_age()

pet1.set_age(-2)

pet1.get_age()


# if you are doing

pet1._age = 5 
# this works, but it will break the principle/idea of ENCAPSULATION

# you could change things from outside, but by convention everyone agrees that
# if anything start with _ like _age ---> treated as an internal 
# and you should avoid touching it directly
````

````python
pet1.introduce()
````

### 2. Inheritance
- a new child class that can reuse the code from an existing parent class
- child classes can add new things or change as well
- 
- Let's create these classes for animal: Dog, Cat, Bird
- All animals - 
    - different behaviours like Sounds
    - 
- 

````python
class Dog(Animal): # dog is a kind of animal
    def __init__(self, name, age):
        super().__init__(name,age, "Dog") # super() refers to the parent class ---> Animal
        # 

    def sound(self):
        print(f"{self.name} says Bhaww Bhaww!!")
````

````python
dog1 = Dog("Tommy", 12)
dog2 = Dog("Bruno", 4)
````

````python
dog1.get_species()

dog1.introduce()
````

### 3. Polymorphism
- same method name working in different ways for different classses
- 
- the same method is poly morphic

````python


class Cat(Animal): # dog is a kind of animal
    def __init__(self, name, age):
        super().__init__(name,age, "Cat") # super() refers to the parent class ---> Animal
        # 
    def sound(self):
        print(f"{self.name} says Meow Meow!!")

class Bird(Animal): # dog is a kind of animal
    def __init__(self, name, age):
        super().__init__(name,age, "Bird") # super() refers to the parent class ---> Animal
        # 
    def sound(self):
        print(f"{self.name} says Kaww Kaww!!")

class Cow(Animal): # dog is a kind of animal
    def __init__(self, name, age):
        super().__init__(name,age, "Quadruple") # super() refers to the parent class ---> Animal
        # 
    def sound(self):
        print(f"{self.name} says Moo Moo!!")
````

````python
cat1 = Cat("Noor", 3)
cat2 = Cat("Bruno", 2)
````

````python
bird1 = Bird("Shebu", 12)
cow1 = Cow("Mitthu", 4)
````

````python
bird1.introduce()

bird1.sound()

cow1.sound()
````

### 4. Abstraction
- in a lot of places in computer sciences
- 
- When all the complexity, messy details are hidden, 
    - only the necessary simple set of commands is given to the user ---> Abstraction

- 

- Here we learn about:
    - Public methods
    - Private methods
- 
- 

````python
import time
import playsound as ps

class CoffeeMachine:
    def __init__(self):
        self.water_temp = 70

    # Private Methods --- Hidden details
    # start with _ to tell users these are private
    def _boil_water(self):
        print("heating water to 112 degree Celsius")
        time.sleep(1)
    
    def _grind_beans(self):
        print("Grinding the Beans ...")
        time.sleep(1)

    def _brew(self):
        print("Ground breans are put into boiling water...")
        time.sleep(1)
        # play a ching sound --- to signal coffee ready
        print("Coffee Ready ...")
        ps.playsound(r"C:\Users\devid\Desktop\Nexperts Academy\AIML-nexperts4\Resources\Ching_cash_register.mp3")


    def _fill_cup(self):
        print("Pouring coffee into cup...")
        # time.sleep(1)
        # play a pouring sound
        ps.playsound(r"C:\Users\devid\Desktop\Nexperts Academy\AIML-nexperts4\Resources\pouring.mp3")


    # public method
    # this is the only method the user is supposed to know/call
    def make_coffee(self):
        # user has pressed the make coffee button
        print("Starting coffee process...")

        self._boil_water()
        self._grind_beans()
        self._brew()
        self._fill_cup()

        print("Here! Please enjoy your coffee...")
````

````python
# user interaction ---> demonstration of abstraction
my_machine = CoffeeMachine()

idrees_Cofee = my_machine.make_coffee()
````
