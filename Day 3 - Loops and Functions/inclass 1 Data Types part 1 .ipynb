### **List**
- A collection of items
- Think of it as: a row of labelled boxes
- we create a list using square brackets `[]`

```python
# use a LIST
roll_numbers = [10,22,9,8]

type(roll_numbers)
print(roll_numbers)
```

```python
# access an element within a variable or a list
roll_numbers[0] # get me the first element
roll_numbers[1] # get me the second element
student4 = roll_numbers[3] # get the element at fourth position
```

If you try to access something that is not within the range of this list:
- IndexError: list index out of range 

```python
# go to third roll number, and change it 
roll_numbers[2] = 99
# we assigned a new value

roll_numbers
```

### **Tuple**
- A collection of items
- Are Immutable - Not changeable
- we create a tuple using round brackets `()`

```python
# use a tuple, save things you do not want to change
registration_nums = (101, 397, 172, 258)
```

```python
registration_nums[0]
registration_nums[3]

# can we change the registration number of the 4th student?
registration_nums[3] = 754

# TypeError ---> tuple does not allow reassignment 
```

### **Range** - more like a method
- Does not help us store data
- That's why we cannot call it a data type
- Otherwise it behaves like sequence data types

```python
type(range)

x = range(5) # inside this are 6 elements - 0 to 5
print(x)

# scriptable - 
```
