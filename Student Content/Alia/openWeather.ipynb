```python
#schedule 
import requests
import pandas as pd

import os
from dotenv import load_dotenv

from datetime import datetime, timedelta  # Cleanest way to import
import time

api_key = os.getenv('API_KEY')

timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

INTERVAL = 60  # 1 minutes in seconds
TOTAL_DURATION = 300  # 5 mnt in seconds

API_KEY = "19f439fb4b3d980c15f3f194dc854037"
CITY_NAME = "London"
BASE_URL = "http://api.openweathermap.org/data/2.5/weather"

# Storage for weather data
weather_records = []

def fetch_weather():
    params = {
        'q': CITY_NAME,
        'appid': API_KEY,
        'units': 'metric'
    }

    try:
        response = requests.get(BASE_URL, params=params)
        response.raise_for_status() 
        data = response.json()
        
        temp = data['main']['temp']
        desc = data['weather'][0]['description']
        display_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        time_now = datetime.now()

        record = {
            "Timestamp": datetime.now(),
            "City": data['name'],
            "Temperature": data['main']['temp'],
            "Description": data['weather'][0]['description'],
            'Humidity': data['main']['humidity'],
            'Pressure': data['main']['pressure']
        }
        
        # Append to list
        weather_records.append(record)

        print(f"[{display_time}] {CITY_NAME}: {temp}°C, {desc} {time_now}")
    except Exception as e:
        print(f"Error fetching data: {e}")

def main():
    print(f"Starting weather monitor for {CITY_NAME}...")
    start_time = datetime.now()
    end_time = start_time + timedelta(seconds=TOTAL_DURATION)
    
    # Check current time against end_time in every iteration
    while datetime.now() < end_time:
        fetch_weather()
        
        # If we have less than 60s left, don't sleep the full minute
        # or just sleep 60 to keep it simple:
        time.sleep(60)

    # Convert the list of dictionaries into a DataFrame
    df = pd.DataFrame(weather_records)
    
    print("\n--- Final Data ---")
    print(df)
    
    # Save to CSV for future use
    df.to_csv("weather_log.csv", index=False)
    print("\nData saved to weather_log.csv")

if __name__ == "__main__":
    main()


```
