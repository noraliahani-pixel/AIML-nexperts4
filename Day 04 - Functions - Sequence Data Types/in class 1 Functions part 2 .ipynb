#### **Paradigms of Programming**
- pattern of programming
- 
#### 1. Imperative
- focus on task completion and state updations

```python
total = 0
for i in range(5):
    total = total + i
print(total)
# Compound sum of all numbers from 0 to 4
```

#### 2. Procedural Programming
- subtype of imperative
- structures program into procedures (functions)
- promotes code reuse

```python
# function/procedure to sum the numbers upto n
def calculate_sum(n):
    total = 0
    for i in range(n + 1):
        total += i            # same as total = total + i
    print(total)

calculate_sum(1000000)
```

#### 3. Object Oriented Programming
- OOP we consider classes and object
- Encapsulation, Inheritance, Polymorphism

#### 4. Declarative Programming
- focuses on outcome, rather than procedure
- commonly used in database querying
- configuration management

#### 5. Functional Programming
- avoids state changes
- Uses:
    - First Class Functions: functions can be passed as arguments or returned from other functions
    - Higher Order Functions: operate on other functions

```python
# function to square a number
def square(x):
    return x * x

square(4)

# i want a square of 5 different numbers
# procedural
sq1 = square(1)
sq2 = square(2)
sq3 = square(3)
sq4 = square(4)
sq5 = square(5)

# functional programming
nums = [1,2,3,4,5]
squares = list(map(square, nums))
```
