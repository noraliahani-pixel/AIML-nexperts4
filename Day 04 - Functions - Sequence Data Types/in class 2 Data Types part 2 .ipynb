##### Slicing 
- we use to access certain items in a list
- 
- list[:] - all elements
- list[start: ] - start from a point upto end
- list[: end] - start from beginning to the end point
- list[start:end] - get a range of elements from start to end
- list[start:end:step] - select a step size
- list[start:end:-step] - negative stepsize reverses the list


Later on, we will see an operation called INDEXING as well

- list[start:end , start:end]


````python
# accessing

new_list[:]
new_list[50:]
new_list[:30]
new_list[20:40]
new_list[::2]
new_list[20:40:2]
new_list[:30:2]
new_list[::-1]
````

### Set

- just a collection of UNIQUE items 
- unordered, you don't have an index
- 
- Does not allow Duplicates
- 

````python
unique_nums = {1,2,3,3,3,3,3}
unique_nums

# set - does not have a index
# not a sequence data type
#unique_nums[0] # TypeError: 'set' object is not subscriptable

#some methods
nums = unique_nums
nums

# add elements to this set
nums.add(4)
nums

# Operations
# UNION and Intersection

a = {1,2,3,4}
b = {3,4,5,6}

a | b # union

a & b # intersection
````

### Dictionary

- Data is stored in Key-Value pairs
- 
- if we have to find an element, we directly target its key
    - so this data structure is fast
- Unordered
- Mutable
- Easter Egg of dictionaries:     
    - syntax matches with JSON - JavaScript Object Notation - to do LLM prompting
- Syntax:
```python
{
    'key':  # key must be a string
    value # value can be any data type
}

````python
# create a dictionary
book = {
    1: 1, # not good practice
    'name': 1984,    # int
    'author': "george orwell", # str
    'year': 1949, # int
    'price': 9.84, # float
    'sales in $K': [200, 143, 104, 86, 32],
    'info': (1984,"George Orwell"),
    'reviews' : {
        'idrees': 'awesome book...',
        'Alia': 'why did our teacher even suggest such a book',
        'Sykran': 'Donald Trump should definitly read this book',
    }
}

book
````

````python
# another way of creating dictionary

student = dict(
    name = "idrees",
    age = 24,
    course = "literature"
)

# if you have a list of tuples

student1 = [
    ('name', "idrees"),
    ('age', 24),
    ('course', ["literature","maths"])
    ]

student1 = dict(student1)


# Using zip function

title = ["name", "age", "course"] # set of keys
student1 = ["idrees",30, "AIML"] # set of values

student2 = ["athirah", 24, "material engg"]

student = dict(zip(title, student1))

student
````

#### Dictionary Operations
- add - add a new keyvalue pair
- update - modify an exisiting one
- delete - remove a keyvalue pair
- .get(), .pop()
- .items(), .keys(), .values()

````python
# access elements in an dictionary
student["name"]

# access something that does not exist, Key Error
# but if you assign value to it, it creates a new element
student["university"] = "Kashmir University"

# Update a value
student['course'] = ["AIML","Math","Astro Physics","Poetry"]

student
````

````python
# Remove details or a key value pair like AGE
del student["course"]


age_stu = student.get('age') # we get the element
student

# if you want to pop out instead
student.pop('age') # can save the value in a variable
name_stu  = student.pop('name')

name_stu
````

##### iterate over dictionary
- dictionary.items() , .keys(), .values()

````python
student.items() # turns dictionary into a list of key-value tuples

student.keys()

student.values()

for i in student.items():
    print(i)
````

````python
for key, value  in student.items():
    print(f"{key} : {value}")
````
